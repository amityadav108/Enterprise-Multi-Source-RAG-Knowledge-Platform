### Import Libraries

```python
import os
import pickle
import pandas as pd
```

### Create Folders

```python
os.makedirs("data/text_docs", exist_ok=True)
os.makedirs("data/csv", exist_ok=True)
```

###  Create Text Documents

```python
employee_handbook = """
Employee Handbook

Employees receive 20 annual leave days.

Remote work is allowed three days per week.

Working hours are 9 AM to 6 PM.

Health insurance benefits are provided.
"""

company_policies = """
Company Policies

Passwords must be changed every 90 days.

VPN access is mandatory.

Multi-factor authentication is required.

Confidential data should not be shared externally.
"""

cloud_migration = """
Cloud Migration Guide

AWS is the primary cloud platform.

Kubernetes is used for deployment.

GitHub Actions manages CI/CD.

Prometheus and Grafana are used for monitoring.
"""
```

###  Save Text Files

```python
with open("data/text_docs/employee_handbook.txt", "w") as f:
    f.write(employee_handbook)

with open("data/text_docs/company_policies.txt", "w") as f:
    f.write(company_policies)

with open("data/text_docs/cloud_migration_guide.txt", "w") as f:
    f.write(cloud_migration)

print("Text files created")
```

### Create employees.csv

```python
employees = pd.DataFrame({
    "EmployeeID":[101,102,103,104,105],
    "Name":["Alice","Bob","Charlie","David","Emma"],
    "Department":["Data Science","ML Engineering","HR","Finance","Cloud"],
    "Experience":[5,3,8,6,4]
})

employees.to_csv("data/csv/employees.csv", index=False)
```

###  Create products.csv

```python
products = pd.DataFrame({
    "ProductID":["P101","P102","P103"],
    "Product":["Laptop","Mouse","Keyboard"],
    "Category":["Electronics","Electronics","Electronics"],
    "Price":[800,25,60]
})

products.to_csv("data/csv/products.csv", index=False)
```

### Create support_tickets.csv

```python
tickets = pd.DataFrame({
    "TicketID":["T001","T002","T003"],
    "Issue":["Login failure","VPN issue","Password reset"],
    "Priority":["High","Medium","Low"]
})

tickets.to_csv("data/csv/support_tickets.csv", index=False)
```

###  Load Everything

```python
from pathlib import Path

documents = []
```

### Load TXT files

```python
for file in Path("data/text_docs").glob("*.txt"):

    with open(file, "r", encoding="utf-8") as f:
        text = f.read()

    documents.append({
        "source": file.name,
        "content": text
    })
```

### Load CSV files

```python
for file in Path("data/csv").glob("*.csv"):

    df = pd.read_csv(file)

    documents.append({
        "source": file.name,
        "content": df.to_string()
    })
```

```python
print("Total documents:", len(documents))
```

### Save documents.pkl

```python
with open("data/documents.pkl", "wb") as f:
    pickle.dump(documents, f)

print("documents.pkl saved")
```
